````python
def initialize_inventory(products):
    inventory={}
    for product in products:
        quantity = int(input(f"Ingrese la cantidad de {product}"))
        inventory[product]=quantity
    return inventory
````

````python
# 2 Modify the `calculate_total_price` function to include error handling.
#  - If the user enters an invalid price (e.g., a negative value or a non-numeric value), 
#    display an error message and ask them to re-enter the price for that product.
#  - Use a try-except block to handle the error and continue prompting the user until a valid price is entered.

'''
def calculate_total_price(customer_orders):
    prices = {product: float(input(f"Enter the price of {product}: ")) for product in customer_orders}
    total_price = sum(prices.values())
    return total_price'''

def calculate_total_price(customer_orders):
    prices = {}
    for product in customer_orders:
        valid_price = False
        while not valid_price:
            try:
                price = float(input(f"Enter the price for {product}: "))
                if price < 0:
                    raise ValueError("Price cannot be negative. Please enter a non-negative value.")
                valid_price = True
            except ValueError as error:
                print(f"Error: {error}")
        prices[product] = price
        total_price = sum(prices.values())
    return total_price
````

````python
# 3. Modify the `get_customer_orders` function to include error handling.
# -If the user enters an invalid number of orders (e.g., a negative value or a non-numeric value), display an error  message and
#   ask them to re-enter the number of orders.
# -If the user enters an invalid product name (e.g., a product name that is not in the inventory), 
#   or that doesn't have stock available, display an error message and ask them to re-enter the product name. 
#       *Hint: you will need to pass inventory as a parameter*
# -Use a try-except block to handle the error and continue prompting the user until a valid product name is entered.

'''def get_customer_orders(products):
    num_orders = int(input("Enter the number of customer orders: "))
    customer_orders = {input("Enter the name of a product that a customer wants to order: ").strip().lower() for _ in range(num_orders) if input("Enter the name of a product that a customer wants to order: ").strip().lower() in products}
    return customer_orders'''


def get_customer_orders(products):
    customer_orders = {}
    while True:
        try:
            num_orders = int(input("Enter the number of different products you want to order: "))
            if num_orders < 0:
                raise ValueError("Number of orders cannot be negative.")
            break
        except ValueError as error:
            print(f"Error: {error}")

    for _ in range(num_orders):
        while True:
            product = input("Enter the product name: ").strip()
            if product not in products:
                print("Invalid product name. Please enter a product that is in the inventory.")
            else:
                break

        while True:
            try:
                quantity = int(input(f"Enter the quantity of {product} you want to order: "))
                if quantity < 0:
                    raise ValueError("Quantity cannot be negative. Please enter a non-negative value.")
                if quantity > products[product]:
                    raise ValueError(f"Not enough stock available for {product}. Available quantity: {products[product]}.")
                break
            except ValueError as error:
                print(f"Error: {error}")

        customer_orders[product] = quantity

    return customer_orders
````
